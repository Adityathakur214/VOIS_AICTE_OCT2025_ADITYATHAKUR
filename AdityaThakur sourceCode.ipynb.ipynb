```python
# Libraries import karein (agar pehle nahi kiye hain)
import matplotlib.pyplot as plt

# Plotting (Histogram)
plt.figure(figsize=(10, 6))

# Hum 'price' column use kar rahe hain. 
# 1000 se upar ki prices ko hum ignore kar rahe hain taki graph saaf dikhe.
df[df['price'] < 1000]['price'].hist(bins=50, edgecolor='black', alpha=0.7)

# Labels aur Title
plt.xlabel("Price ($)")
plt.ylabel("Number of Listings")
plt.title("Price Distribution (Listings under $1000)")

# Plot dikhana
plt.show()
```

```python
# 1. Top 10 Hosts ka data nikalna
# 'host_name' aur 'calculated host listings count' columns ka upyog karein.
# Hum maan rahe hain ki calculated host listings count ka column name yahi hai.
host_data = df.groupby(['host name'])['calculated host listings count'].count().sort_values(ascending=False).head(10).reset_index()

# 2. Plotting ke liye libraries import karein 
import matplotlib.pyplot as plt

# 3. Plotting (Matplotlib Bar Chart)
# Chart size set karna
plt.figure(figsize=(12, 6)) 

# plt.bar() ka upyog karke plot banana
plt.bar(host_data['host name'], host_data['calculated host listings count'], color='skyblue', edgecolor='black')

# 4. Labels aur Title set karna
plt.xlabel("Host Name")
plt.ylabel("Calculated Host Listings Count")
plt.title("Top 10 Hosts by Calculated Host Listings Count")

# X-axis labels ko rotate karna taki host ke naam aasani se padh saken
plt.xticks(rotation=45, ha='right')
plt.ylim(0, host_data['calculated host listings count'].max() * 1.1) # Y-axis limit set karna

# Plot ko tightly adjust karna
plt.tight_layout() 

# Plot dikhana
plt.show()
```

```python
# Sahi column names ka upyog: 'neighbourhood group', 'room type', aur 'review rate number'

# 1. Groupby operation
# df.groupby() ka upyog karke do columns par grouping karna aur review rate ka average nikalna
ARRN = df.groupby(['neighbourhood group', 'room type'])['review rate number'].mean().to_frame()

# 2. Output ko dekhna
print("Average Review Rate Number:")
print(ARRN)

# Note: Yeh code ek table output karega jaisa ki aapke image mein hai, plot nahi. 
# Agar aapko plot chahiye, toh aapko naye cell mein ise visual karna hoga (jaise ki heatmap).
```

```python
# Sahi column names ka upyog: 'neighbourhood group', 'room type', aur 'review rate number'
# Data nikal kar ARR (Average Review Rate) DataFrame banaya ja raha hai
ARRN = df.groupby(['neighbourhood group', 'room type'])['review rate number'].mean().to_frame().reset_index()

# Heatmap ke liye data ko pivot karna
heatmap_data = ARRN.pivot("neighbourhood group", "room type", "review rate number")
```

```python
# Libraries import karein (agar pehle nahi kiye hain)
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting (Heatmap)
plt.figure(figsize=(12, 7)) 

# Seaborn heatmap function ka upyog
sns.heatmap(
    heatmap_data, # Pivot kiya hua data use karein
    annot=True, # Cells mein numbers dikhane ke liye
    fmt=".2f", # Do decimal places tak numbers dikhana
    cmap="YlGnBu", # Colour scheme set karna
    linewidths=.5, # Cells ke beech ki line
    cbar_kws={'label': 'Average Review Rate Number'} # Colour bar ka label
)

# Labels aur Title set karna
plt.xlabel("Room Type")
plt.ylabel("Neighbourhood Group")
plt.title("Average Review Rate for each Room/Property Type in each Neighbourhood Group")

# Plot dikhana
plt.show()
```

```python
# Libraries import karein (agar pehle nahi kiye hain)
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Data Preparation
# Grouping aur average review rate nikalna (ARRN creation)
ARRN = df.groupby(['neighbourhood group', 'room type'])['review rate number'].mean().to_frame().reset_index()

# Heatmap ke liye data ko pivot karna (yahi line pichle baar miss ho gayi thi)
heatmap_data = ARRN.pivot("neighbourhood group", "room type", "review rate number")

# 2. Plotting (Heatmap)
plt.figure(figsize=(12, 7)) 

sns.heatmap(
    heatmap_data, 
    annot=True, 
    fmt=".2f", 
    cmap="YlGnBu", 
    linewidths=.5, 
    cbar_kws={'label': 'Average Review Rate Number'}
)

# Labels aur Title set karna
plt.xlabel("Room Type")
plt.ylabel("Neighbourhood Group")
plt.title("Average Review Rate for each Room/Property Type in each Neighbourhood Group")

plt.show()
```

```python
# Libraries import karein
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Data Preparation (Corrected Pivot Syntax)
# Grouping aur average review rate nikalna (ARRN creation)
ARRN = df.groupby(['neighbourhood group', 'room type'])['review rate number'].mean().to_frame().reset_index()

# **ERROR FIXED HERE:** keyword arguments (index, columns, values) use kiye gaye hain
heatmap_data = ARRN.pivot(
    index="neighbourhood group", 
    columns="room type", 
    values="review rate number"
)

# 2. Plotting (Heatmap)
plt.figure(figsize=(12, 7)) 

sns.heatmap(
    heatmap_data, 
    annot=True, 
    fmt=".2f", 
    cmap="YlGnBu", 
    linewidths=.5, 
    cbar_kws={'label': 'Average Review Rate Number'}
)

# Labels aur Title set karna
plt.xlabel("Room Type")
plt.ylabel("Neighbourhood Group")
plt.title("Average Review Rate for each Room/Property Type in each Neighbourhood Group")

plt.show()
```
